```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Values.csv", delimiter = ",", skiprows=1)
channel = data.iloc[:,0]
intensity = data.iloc[:,1]
```

```python
plt.plot(channel, intensity)
plt.xlabel("Channel")
plt.ylabel("Relative intensity")
plt.grid()
plt.show()
```

- 56 == 517 nm
- 107 == 656 nm

```python
m = (656-517)/(107.5-56)
print(m)
```

```python
wavelength = m*channel + 366
```

```python
plt.figure(dpi = 500)
plt.plot(wavelength, intensity, "-", lw = 1)

##Hydrogen
plt.vlines(434, 0, 20.5, ls = "-.", color = "mediumblue", lw = 0.5, label = "H_gamma (434 nm)")
plt.vlines(486, 0, 34, ls = "-.", color = "green", lw = 0.5, label = "H_beta (486 nm)")
plt.vlines(656, 0, 35.3, ls = "-.", color = "red", lw = 0.5,  label = "H_alpha (656 nm)")

##Helium
plt.vlines(588, 0, 41.5, ls = "-", color = "gold", lw = 0.5, label = "He (588 nm)")

##Oxygen
plt.vlines(615, 0, 41.5, ls = ":", color = "orange", lw = 0.5, label = "O (615 nm)")

##Iron
plt.vlines(517, 0, 34, ls = ":", color = "k", lw = 0.5, label = "Fe (517 nm)")



plt.xlabel("Wavelength")
plt.ylabel("Relative intensity")
plt.legend(fontsize = 5)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import time
```

```python
def pot(x):
    for i in range(-101, 107, 3):
        if (i + 0.6 < x < (i + 3.4)):
            return 0
    return 5
```

```python
h = 0.1
x = np.arange(-101, 107, h)
L = len(x)
h_bar = 1
m = 1
```

```python
value_potential = np.zeros(L)
for i in range(L):
    value_potential[i] = pot(x[i])


plt.plot(x, value_potential)
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('Potential')
plt.grid()
plt.show()
```

```python
##Define H matrix
H = np.zeros((L-1,L-1))
for i in range(L-1):
    H[i][i] = 2*h_bar**2 / (2*m*h**2) + pot(x[i+1])
    if i > 0:
        H[i][i-1] = -1*h_bar**2 / (2*m*h**2)
    if i < L-2:
        H[i][i+1] = -1*h_bar**2 / (2*m*h**2)
```

```python
eigv, eigf = sp.linalg.eigh(H)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x[1:], (-1)**0 *eigf[:,0], label = f"Energy = {eigv[0]:0.6f}")
plt.xlim(-6,6)
# plt.ylim(0.020, 0.040)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Ground State Probability Density')
plt.grid()
plt.legend()
plt.show()
```

```python
for i in range(0, 5):
    plt.figure(figsize = (10,5), dpi = 100)
    plt.plot(x[1:], (eigf[:,i])**2, label = f"Energy = {eigv[i]:0.6f}")
    plt.xlabel("Distance")
    plt.ylabel(r"$|{\psi}(x)|^2$")
    plt.title("Wavefunctions")
    plt.legend()
    plt.grid()
    plt.show()
```
